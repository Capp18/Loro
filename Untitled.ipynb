```python
def statistics1(df):
    
'''
Questa funzione prende una tabella e calcola statistiche aggregate rispetto a una colonna, in questo caso SK_ID_BUREAU
'''

    df1 = df #creiamo un nuovo df in memoria che possiamo modificare senza intervenire su quello originale
    
    df1  = df1.drop(columns = ['SK_ID_BUREAU'])     
    set_ID = list(set(df1['SK_ID_CURR'].tolist()))  # lista di tutti gli ID senza duplicati
    
    result = pd.DataFrame()
    
    cols_count = []; cols_max = []; cols_min = []; cols_sum = []; cols_mean = []
    
    
    for column_name, _ in df1.iteritems():
        
        #non consideriamo le colonne con info in stringhe
            
        if np.issubdtype(df1[column_name].dtype , np.number) == False:  
            df1 = df1.drop(columns = [column_name])
            
        else: #se sono numeriche 
            
            if column_name == 'SK_ID_CURR':
                cols_count.append(str(column_name))
                cols_max.append(str(column_name))
                cols_min.append(str(column_name))
                cols_sum.append(str(column_name))
                cols_mean.append(str(column_name))
            
            elif column_name != 'SK_ID_CURR':
                    #Inizializza le colonne del df
                    cols_count.append(str(column_name) +'_count')
                    cols_max.append(str(column_name) +'_max')
                    cols_min.append(str(column_name) +'_min')
                    cols_sum.append(str(column_name) +'_sum')
                    cols_mean.append(str(column_name) +'_media')
    
    cols = cols_count + cols_max[1:] + cols_min[1:] + cols_sum[1:] + cols_mean[1:]
    
    df = df1.groupby('SK_ID_CURR',as_index=False)
    
    count = df.count()
    
    Max = df.max()
    
    Min = df.min()
    
    Sum = df.sum()
    
    mean = df.mean()
    
    
    df1 = pd.merge(count,Max, on = 'SK_ID_CURR')
    df1 = pd.merge(df1,Min, on = 'SK_ID_CURR')
    df1 = pd.merge(df1,Sum, on = 'SK_ID_CURR' )
    df1 = pd.merge(df1,mean, on = 'SK_ID_CURR' )
    
    df1.columns = [cols]


    return df1

def statistics2(df):
    df1 = df #creiamo un nuovo df in memoria che possiamo modificare senza intervenire su quello originale
    
    #df1  = df1.drop(columns = ['SK_ID_BUREAU'])     
    set_ID = list(set(df1['SK_ID_BUREAU'].tolist()))  # lista di tutti gli ID senza duplicati
    
    result = pd.DataFrame()
    
    cols_count = []; cols_max = []; cols_min = []; cols_sum = []; cols_mean = []
    
    
    for column_name, _ in df1.iteritems():
        
        #non consideriamo le colonne con info in stringhe
            
        if np.issubdtype(df1[column_name].dtype , np.number) == False:  
            df1 = df1.drop(columns = [column_name])
            
        else: #se sono numeriche 
            
            if column_name == 'SK_ID_BUREAU':
                cols_count.append(str(column_name))
                cols_max.append(str(column_name))
                cols_min.append(str(column_name))
                cols_sum.append(str(column_name))
                cols_mean.append(str(column_name))
            
            elif column_name != 'SK_ID_BUREAU':
                    #Inizializza le colonne del df
                    cols_count.append(str(column_name) +'_count')
                    cols_max.append(str(column_name) +'_max')
                    cols_min.append(str(column_name) +'_min')
                    cols_sum.append(str(column_name) +'_sum')
                    cols_mean.append(str(column_name) +'_media')
    
    cols = cols_count + cols_max[1:] + cols_min[1:] + cols_sum[1:] + cols_mean[1:]
    
    df = df1.groupby('SK_ID_BUREAU',as_index=False)
    
    count = df.count()
    
    Max = df.max()
    
    Min = df.min()
    
    Sum = df.sum()
    
    mean = df.mean()
    
    
    df1 = pd.merge(count,Max, on = 'SK_ID_BUREAU')
    df1 = pd.merge(df1,Min, on = 'SK_ID_BUREAU')
    df1 = pd.merge(df1,Sum, on = 'SK_ID_BUREAU' )
    df1 = pd.merge(df1,mean, on = 'SK_ID_BUREAU' )
    
    df1.columns = [cols]


    return df1
```

```python
def stats(df,column):
    
    '''
    Questa funzione prende una tabella e una colonna e calcola statistiche aggregate rispetto 
    alla colonna indicarta, ritornando un  df che non ha più una struttuta a multindex
    '''

    df1 = df #creiamo un nuovo df in memoria che possiamo modificare senza intervenire su quello originale
    
    if column =='SK_ID_CURR':
        df1  = df1.drop(columns = ['SK_ID_BUREAU']) #chek valido per il caso di bureau.csv per non fare statistiche
                                                    #sulla colonna di matricole di SK_ID_BUREAU
        
    #set_ID = list(set(df1[column].tolist()))  # lista di tutti gli ID senza duplicati
    #result = pd.DataFrame()
    
    cols_count = []; cols_max = []; cols_min = []; cols_sum = []; cols_mean = []
    
    
    for column_name, _ in df1.iteritems():
        
        #non consideriamo le colonne con info in stringhe
            
        if np.issubdtype(df1[column_name].dtype , np.number) == False:  
            df1 = df1.drop(columns = [column_name])
            
        else: #se sono numeriche
            
            if column_name != column:
                cols_count.append(str(column_name) +'_count')
                cols_max.append(str(column_name) +'_max')
                cols_min.append(str(column_name) +'_min')
                cols_sum.append(str(column_name) +'_sum')
                cols_mean.append(str(column_name) +'_media')
                                 
    print((cols_count))
    cols =[column] + cols_count + cols_max + cols_min + cols_sum + cols_mean
    
    df = df1.groupby(column,as_index=False)
    
    count = df.count()
   # print(count.columns)
    
    Max = df.max()
    #print(Max.columns)
    
    Min = df.min()
  #  print(Min.columns)

    Sum = df.sum()
    
    mean = df.mean()
    
    
    df1 = pd.merge(count,Max, on = column)
    #print(df1.columns)
    df1 = pd.merge(df1,Min, on = column)
    #print(df1.columns)
    df1 = pd.merge(df1,Sum, on = column )
    #print(df1.columns)
    df1 = pd.merge(df1,mean, on = column )
    #print(df1.columns)
    
    df1.columns = [cols]
    df1.columns = ['_'.join(col).strip() for col in df1.columns.values]  #to flatten the multindex


    return df1


def term_freq(df,column):        

    cols_sum = []; cols_mean = []
    
    cat = pd.get_dummies(df.select_dtypes('object')) #nuovo dataframe solo con colonne di tipo ogetto
    
    for column_name, _ in cat.iteritems():
        if column_name == column:
            cols_sum.append(column_name)
        else:
            cols_sum.append(str(column_name)+'_sum')
            cols_mean.append(str(column_name)+'_mean')
            
    cols = [column]+ cols_sum + cols_mean
    
    cat[column] = df[column]
    
    group = cat.groupby(column,as_index=False)
    
    Sum = group.sum()
    
    mean = group.mean()
    
    df1 = pd.merge(Sum,mean, on = column)
    df1.columns = [cols]
    df1.columns = ['_'.join(col).strip() for col in df1.columns.values]  #to flatten the multindex

    return df1
```
